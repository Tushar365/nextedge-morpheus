```python
import pandas as pd
from openai import OpenAI

# Initialize the OpenAI client
client = OpenAI(
    api_key=AIML_API_KEY,
    base_url="https://api.aimlapi.com"
)

# Define the prompt and its category
prompt_and_category = {
    "prompt": "Generate another example of a positive social media post. Just provide the text of the post, nothing else.",
    "category": "positive"
}

# List to store the synthetic data
synthetic_data = []

# Generate 10 different samples for the same prompt
for count in range(5):
    print("generating ",count)
    response = client.chat.completions.create(
        model="meta-llama/Meta-Llama-3.1-70B-Instruct-Turbo",
        messages=[
            {
                "role": "user",
                "content": prompt_and_category["prompt"]
            },
        ],
        temperature=0.8,  # Controls randomness: Higher values lead to more diverse outputs
        top_p=0.9         # Controls diversity via nucleus sampling: only consider top_p probability mass
    )

    generated_text = response.choices[0].message.content

    synthetic_data.append({
        "prompt": prompt_and_category["prompt"],
        "generated_text": generated_text,
        "category": prompt_and_category["category"]
    })

# Convert the synthetic data into a DataFrame for easier handling
df_synthetic_data = pd.DataFrame(synthetic_data)

# Save the DataFrame to a CSV file
df_synthetic_data.to_csv("synthetic_data_10_samples_with_category.csv", index=False)

print("Synthetic data generation complete. Data saved to 'synthetic_data_10_samples_with_category.csv'.")
```

```python
import pandas as pd
from openai import OpenAI

# Initialize the OpenAI client
client = OpenAI(
    api_key=AIML_API_KEY,
    base_url="https://api.aimlapi.com"
)

# Define prompts and their associated categories
prompts_and_categories = [
    {"prompt": "Generate an example of a positive social media post. Just provide the text of the post, nothing else.:", "category": "positive"},
    {"prompt": "Create an example of a spammy comment on a video. Just provide the text of the post, nothing else.:", "category": "spam"},
    # {"prompt": "Write an inappropriate comment that might be flagged:", "category": "inappropriate"},
    {"prompt": "Compose a neutral statement about technology. Just provide the text of the post, nothing else.:", "category": "neutral"},
    {"prompt": "Generate an example of a misleading news headline. Just provide the text of the post, nothing else.:", "category": "misleading"}
]

# List to store the synthetic data
synthetic_data = []

# Generate 100 posts for each prompt and category
for item in prompts_and_categories:
    for count in range(100):
        print(f"Generating sample {count + 1} for category '{item['category']}'")
        response = client.chat.completions.create(
            model="meta-llama/Meta-Llama-3.1-70B-Instruct-Turbo",
            messages=[
                {
                    "role": "user",
                    "content": item["prompt"]
                },
            ],
            temperature=0.8,  # Controls randomness: Higher values lead to more diverse outputs
            top_p=0.9         # Controls diversity via nucleus sampling: only consider top_p probability mass
        )

        generated_text = response.choices[0].message.content

        synthetic_data.append({
            "generated_text": generated_text,
            "category": item["category"]
        })

        # Convert the synthetic data into a DataFrame with only posts and categories
        df_synthetic_data = pd.DataFrame(synthetic_data, columns=["generated_text", "category"])
        
        # Save the DataFrame to a CSV file
        df_synthetic_data.to_csv("synthetic_data_500_samples_with_categoriesv2.1.csv", index=False)
        
        print("Synthetic data generation complete. Data saved to 'synthetic_data_500_samples_with_categoriesv2.csv'.")
```
